```python
import requests
from bs4 import BeautifulSoup

url_target = 'https://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance'

# 在network找到post data
post_data_str = """method: search
searchMethod: true
searchTarget: ATM
orgName: 
orgId: 
hid_1: 1
tenderName: 
tenderId: 
tenderStatus: 5,6,20,28
tenderWay: 
awardAnnounceStartDate: 109/09/25
awardAnnounceEndDate: 109/09/25
proctrgCate: 
tenderRange: 
minBudget: 
maxBudget: 
item: 
hid_2: 1
gottenVendorName: 
gottenVendorId: 
hid_3: 1
submitVendorName: 
submitVendorId: 
location: 
execLocationArea: 
priorityCate: 
isReConstruct: 
btnQuery: 查詢"""

# 放入字典
post_data = {}
for row in post_data_str.split('\n'):
    post_data[row.split(': ')[0]] = row.split(': ')[1]

headers = {'User-Agent' : "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"}

res = requests.post(url_target, headers=headers, data=post_data)
soup = BeautifulSoup(res.text, 'html.parser')
# print(soup)

# 列出表格標題和內容
title = soup.select('div[id="print_area"] td[align="left"]')

for n, i in enumerate(title):
    if (n+1) % 4 == 2:
        print(i.u.text)
    else:
        print(i.text)
    if (n+1) % 4 == 0:
        print('=' * 20)
```
